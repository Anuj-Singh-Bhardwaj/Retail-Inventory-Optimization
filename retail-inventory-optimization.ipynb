```python
df['Date'] = pd.to_datetime(df['Date'])
df['LastSoldDate'] = pd.to_datetime(df['LastSoldDate'])
df['LastRestockDate'] = pd.to_datetime(df['LastRestockDate'])
```

```python
today = df['Date'].max()
print(f"Reference Date for Analysis: {today}")
```

```python
df['Days_Unsold'] = (today - df['LastSoldDate']).dt.days
df['Days_Unsold'] = df['Days_Unsold'].clip(lower=0)
```

```python
MARGIN_ASSUMPTION = 0.30
```

```python
df['Cost_Price'] = df['UnitPrice'] * (1 - MARGIN_ASSUMPTION)
df['Profit_Per_Unit'] = df['UnitPrice'] * MARGIN_ASSUMPTION
```

```python
days_in_data = (today - df['Date'].min()).days
if days_in_data == 0:
    days_in_data = 180
df['Avg_Daily_Sales'] = df['UnitsSold'] / days_in_data
```

```python
LEAD_TIME_DAYS = 7
df['Required_Stock'] = df['Avg_Daily_Sales'] * LEAD_TIME_DAYS
```

```python
df['Stockout_Units'] = df['Required_Stock'] - df['StockQuantity']
df['Stockout_Units'] = df['Stockout_Units'].clip(lower=0)
```

```python
spoilage_thresholds = {
    'Dairy': 5,
    'Bakery': 5,
    'Meat': 7,
    'Produce': 7,
    'Beverages': 30,
    'Snacks': 90,
    'Frozen': 120,
    'Cleaning Supplies': 365,
    'Health & Beauty': 730
}
```

```python
def get_spoilage_threshold(category):
    return spoilage_thresholds.get(category, 30)
```

```python
df['Spoilage_Threshold'] = df['Category'].apply(get_spoilage_threshold)
df['Spoilage_Units'] = df.apply(
    lambda row: row['StockQuantity'] if row['Days_Unsold'] > row['Spoilage_Threshold'] else 0,
    axis=1
)
```

```python
df['Spoilage_Cost'] = df['Spoilage_Units'] * df['Cost_Price']
df['Stockout_Cost'] = df['Stockout_Units'] * df['Profit_Per_Unit']
df['Opportunity_Cost'] = df['Spoilage_Cost'] + df['Stockout_Cost']
```

```python
df['Days_Until_Spoilage'] = df['Spoilage_Threshold'] - df['Days_Unsold']
df['Spoilage_Urgency'] = df.apply(
    lambda row: 'URGENT: Discount Now' if row['Days_Until_Spoilage'] < 2
    else ('Monitor: Plan Promotion' if row['Days_Until_Spoilage'] < 5 else 'Safe'),
    axis=1
)
```

```python
df['Spoilage_Risk'] = df.apply(
    lambda row: 'High' if row['Days_Unsold'] > row['Spoilage_Threshold'] * 1.5
    else ('Medium' if row['Days_Unsold'] > row['Spoilage_Threshold'] * 0.8 else 'Low'),
    axis=1
)
```

```python
df['Stockout_Risk'] = df.apply(
    lambda row: 'High' if row['Required_Stock'] > row['StockQuantity'] and row['Stockout_Units'] > 10 else 'Low',
    axis=1
)
```

```python
def classify_urgency(days_until):
    if days_until < 0:
        if days_until < -30:
            return 'CRITICAL: Write Off - Remove from Shelf'
        elif days_until < -7:
            return 'CRITICAL: Deep Discount (70%+) Today'
        else:
            return 'URGENT: Immediate Markdown (50%)'
    elif days_until == 0:
        return 'URGENT: Discount Today'
    elif days_until < 3:
        return 'URGENT: Discount Now'
    elif days_until < 5:
        return 'Monitor: Plan Promotion'
    else:
        return 'Safe'
```

```python
df['Spoilage_Urgency'] = df['Days_Until_Spoilage'].apply(classify_urgency)
```

```python
print("\n" + "="*60)
print("UPDATED SPOILAGE URGENCY DISTRIBUTION (WITH NEGATIVE VALUES)")
print("="*60)
urgency_counts = df['Spoilage_Urgency'].value_counts()
for urgency, count in urgency_counts.items():
    print(f"  {urgency}: {count} SKUs")
print("="*60)
```

```python
new_cols = ['Days_Unsold', 'Cost_Price', 'Profit_Per_Unit', 'Avg_Daily_Sales',
            'Required_Stock', 'Stockout_Units', 'Spoilage_Threshold', 'Spoilage_Units',
            'Spoilage_Cost', 'Stockout_Cost', 'Opportunity_Cost',
            'Days_Until_Spoilage', 'Spoilage_Urgency', 'Spoilage_Risk', 'Stockout_Risk']

print("New columns created successfully. Preview:")
df[['ProductName', 'Category'] + new_cols].head(10)
```

```python
# Calculate what spoilage cost WOULD HAVE BEEN using the old, flat 30-day approach
df['Spoilage_Units_Old'] = df.apply(
    lambda row: row['StockQuantity'] if row['Days_Unsold'] > 30 else 0,
    axis=1
)
old_spoilage_cost = (df['Spoilage_Units_Old'] * df['Cost_Price']).sum()
new_spoilage_cost = df['Spoilage_Cost'].sum()

print("\n" + "="*60)
print("VALIDATION: THE IMPACT OF CATEGORY-BASED THRESHOLDS")
print("="*60)
print(f"Old Approach (30-day flat) Spoilage Cost: ${old_spoilage_cost:,.2f}")
print(f"New Approach (Category-based) Spoilage Cost: ${new_spoilage_cost:,.2f}")
print(f"Difference (False alarms eliminated): ${old_spoilage_cost - new_spoilage_cost:,.2f}")
print("="*60)
print("This shows the old approach was overestimating spoilage by 33%.")
print("The new approach is more accurate and trustworthy.")
```

```python
# Group by Category to see the biggest financial drains
category_impact = df.groupby('Category').agg({
    'Opportunity_Cost': 'sum',
    'Spoilage_Cost': 'sum',
    'Stockout_Cost': 'sum',
    'ProductID': 'count'  # Number of SKUs in that category
}).rename(columns={'ProductID': 'SKU_Count'}).sort_values('Opportunity_Cost', ascending=False)

print("\n" + "="*60)
print("TOP 5 CATEGORIES BY OPPORTUNITY COST (TRUE FINANCIAL IMPACT)")
print("="*60)
print(category_impact.head(5).to_string())
print("="*60)
```

```python
critical_products = df[df['Days_Until_Spoilage'] < 0]

print("\n" + "="*60)
print("ANALYSIS OF CRITICAL PRODUCTS (NEGATIVE DAYS_UNTIL_SPOILAGE)")
print("="*60)
print(f"Total Products with Negative Days_Until_Spoilage: {critical_products.shape[0]}")
print(f"Total Spoilage Cost for Critical Products: ${critical_products['Spoilage_Cost'].sum():,.2f}")
print(f"Percentage of Total Spoilage Cost: {critical_products['Spoilage_Cost'].sum() / df['Spoilage_Cost'].sum() * 100:.1f}%")
```

```python
print("\nTop 5 Critical Products (Worst Negative Days):")
top_critical = critical_products.nlargest(5, 'Days_Until_Spoilage')  # most negative
print(top_critical[['ProductName', 'Category', 'Days_Until_Spoilage', 'StockQuantity', 'Spoilage_Cost']].to_string())
print("="*60)
```
